# output_median_optimized_2

```python
import os as os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

np.random.seed(0)
labels = ['ideal','noise','outlier_1','outlier_2','outlier_3','outlier_4','outlier_5','outlier_6', 'outlier_7']
noise = np.random.normal(0,1,1920)
outlier = 10
x = np.linspace(0, 1919, 1920)
```

## line

```python
plt.rcParams['figure.figsize'] = [21, 7]

# Get Data
y_in = np.zeros((1920, 9))
y_out = np.zeros((1920, 9))
for i in range(0, 9):
    y_in[:,i] = np.loadtxt('./input/line/'+labels[i]+'.dat')
    y_out[:,i] = np.loadtxt('./output_median_optimized_2/line/'+labels[i]+'.dat')

# Plot
fig, axs = plt.subplots(3, 3, sharex=True, sharey=True)
fig.suptitle('Line')
for i, ax in enumerate(axs.flat):
    ax.set(title=labels[i])
    ax.plot(x, y_in[:,i], '.')
    ax.plot(x, y_out[:,i], '.')
    ax.legend(['input', 'output_median_optimized_2'], loc='lower right')
    if i in [0,3,6]:
        ax.set(ylabel='Z (Pixels)')
    if i in [6,7,8]:
        ax.set(xlabel="X (Pixels)")
plt.show()
```
