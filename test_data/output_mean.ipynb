```python
import os as os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

labels = ['noise','outlier_1','outlier_2','outlier_3','outlier_4','outlier_5']

x = np.linspace(0, 1919, 1920)
```

## line

```python
plt.rcParams['figure.figsize'] = [21, 7]

# Get Data
y_in = np.zeros((1920, 6))
y_out = np.zeros((1920, 6))
for i in range(0, 6):
    y_in[:,i] = np.loadtxt('./input/line/'+labels[i]+'.dat')
    y_out[:,i] = np.loadtxt('./output_mean/line/'+labels[i]+'.dat')
    
# Plot
fig, axs = plt.subplots(2, 3, sharex=True, sharey=True)
fig.suptitle('Line')
for i, ax in enumerate(axs.flat):
    ax.set(title=labels[i])
    ax.plot(x, y_in[:,i], '.')
    ax.plot(x, y_out[:,i], '.')
    ax.legend(['input', 'output_mean'], loc='lower right')
    if i in [0,3]:
        ax.set(ylabel='Z (Pixels)')
    if i in [3,4,5]:
        ax.set(xlabel="X (Pixels)")
plt.show()
```
